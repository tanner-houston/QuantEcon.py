```python
#Version 1
import numpy as np
import matplotlib.pyplot as plt

e_values = np.random.randn(100)
plt.plot(e_values)
plt.show()
```

```python
#Alternative version 1 (using a for loop)
ts_length = 100
e_values = [] #empty list for values

for i in range(ts_length):
    e = np.random.randn()
    e_values.append(e)

plt.plot(e_values)
plt.show()
```

```python
ts_length = 100
e_values = []

i = 0
while i < ts_length:
    e = np.random.randn()
    e_values.append(e)
    i = i + 1

plt.plot(e_values)
plt.show()
```

```python
## Bank account over time

r = 0.025    # interest rate
T = 50       # end date
b = np.empty(T+1)   # empty Numpy array to store b_t
b[0] = 10   # Account starts with 10

for t in range(T):
    b[t+1] = (1+r)*b[t]

plt.plot(b, label = "Bank Balance")
plt.legend()
plt.show()
```

```python
#Exercises
#1) simulate a correlated time series

alpha = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    e = np.random.randn()
    x[t+1] = alpha*x[t] + e

plt.plot(x, label = "Time Series")
plt.legend()
plt.show()
```

```python
#Exercises
#2) Three correlated time series

alpha = [0, 0.8, 0.98]
x = np.empty(T+1)

for a in alpha:
    for t in range(T):
        x[t+1] = a * x[t] + np.random.randn()
    plt.plot(x, label = f'$\\alpha = {a}$')
    
plt.legend()
plt.show()
```

```python
#Exercises
#3) adding the absolute value for x

alpha = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    e = np.random.randn()
    x[t+1] = alpha*abs(x[t]) + e

plt.plot(x, label = f'$\\alpha = {alpha}$')
plt.legend()
plt.show()
```

```python
#Exercises
#4) Use conditioning instead of the absolute value function 

alpha = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    if x[t] < 0:
        abs_x = -x[t] 
    else:
        abs_x = x[t]
    x[t+1] = alpha * abs_x + np.random.randn()

plt.plot(x)
plt.show()
```

```python
#Exercises
#5) Compute an approximation of π using Monte Carlo

n = 1000000  #sample size for monte carlo

count = 0
for i in range(n):

    #drawing random positions on the square
    u, v = np.random.uniform(), np.random.uniform()

    #check whether the points falls within the boundary
    #of the unit circle centered at (0.5, 0.5)
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)

    #if it falls within the inscribed circle,
    #add it to the count
    if d < 0.5:
        count += 1

area_estimate = count / n  

print(area_estimate * 4) #dividing by radius**2
```
