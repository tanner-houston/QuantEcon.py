```python
#Version 1
import numpy as np
import matplotlib.pyplot as plt

e_values = np.random.randn(100)
plt.plot(e_values)
plt.show()
```

```python
#numpy attributes
print(np.log(4))
print(np.sqrt(4))
```

```python
#Alternative version 1 (using a for loop)
ts_length = 100
e_values = [] #empty list for values

for i in range(ts_length):
    e = np.random.randn()
    e_values.append(e)

plt.plot(e_values)
plt.show()
```
